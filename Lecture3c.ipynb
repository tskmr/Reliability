```r
a <- -1; b <- 1;
y <- function(u){ exp(u*(b-a)+a) }
u <- runif(10000)
y(u)
```

```r
library(tidyverse)
data.frame(y=y(u)) %>% ggplot(aes(y)) + geom_histogram()
```

```r
data.frame(y=y(u)) %>% ggplot(aes(y, ..density..)) + geom_histogram()
```

```r
exp(a); exp(b)
```

```r
fy = function(y){ 1/((b-a)*y) }
data.frame(y=y(u), dy=fy(y(u))) %>% ggplot() + geom_histogram(aes(y, ..density..)) + geom_line(aes(y, dy), col=2)
```

```r
fy = function(y){ 1/((b-a)*y) }
x <- seq(exp(a), exp(b), len=10000)
data.frame(x, y=y(u), dy=fy(x)) %>% 
  ggplot() + geom_histogram(aes(y, ..density..)) + 
    geom_line(aes(x, dy), col=2)
```

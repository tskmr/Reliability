# 確率プロット

## 正規分布

```r
t <- seq(from=0, to=100, len=100)
```

```r
t
```

```r
p <- pnorm(t, mean=40, sd=10)
p
```

```r
data.frame(t, p) %>% ggplot() + geom_line(aes(t, p))
```

```r
q <- qnorm(p, mean=0, sd=1)
q
```

```r
data.frame(t, p, q) %>% ggplot() + geom_line(aes(t, q))
```

```r
pnorm((log(t)-40)/10, mean=0, sd=1)
pnorm(log(t), mean=40, sd=10)
```

## 対数正規分布

```r
t <- seq(from=0, to=100, len=100)
p <- pnorm((log(t)-40)/10, mean=0, sd=1)
q <- qnorm(p)
data.frame(t, p, q) %>% ggplot() + geom_line(aes(log(t), q))
```

## 最小極値分布

```r
psev <- function(z){ 1- exp(-exp(z)) }
qsev <- function(p){ log(-log(1-p)) }

t <- seq(from=0, to=100, len=100)
p <- psev((t-40)/10)
q <- qsev(p)
data.frame(t, p, q) %>% ggplot() + geom_line(aes(t, q))
```

## ワイブル分布

```r
psev <- function(z){ 1- exp(-exp(z)) }
qsev <- function(p){ log(-log(1-p)) }

t <- seq(from=0, to=100, len=100)
p <- psev((log(t)-40)/10)
q <- qsev(p)
data.frame(t, p, q) %>% ggplot() + geom_line(aes(log(t), q))
```

# ノンパラメトリック推定

```r
library(tidyverse)
t = c( 94,  96,  99,  99, 104, 108, 112, 114, 117, 117, 118, 121, 121, 123, 129,
131, 133, 135, 136, 139, 139, 140, 141, 141, 143, 144, 149, 149, 152, 153,
159, 159, 159, 159, 162, 168, 168, 169, 170, 170, 171, 172, 173, 176, 177,
180, 180, 184, 187, 188, 189, 190, 196, 197, 203, 205, 211, 213, 224, 226,
227, 256, 257, 269, 271, 274, 291)
n = length(t)
i = seq_along(t)
d = data.frame(i, t)
d
```

```r
d %>%
ggplot(aes(0, i))+
geom_segment(aes(xend=t, yend=i), show.legend = F)+
geom_text(aes(x=t, label="x"), show.legend = F)+
xlab(expression(paste("Thousandas of Cycles", " ", ("" %*% 10^3))))+
ylab("Specimens")+
scale_y_reverse()
```

```r
# ノンパラメトリック推定
d = d %>% mutate(p = (i-.5)/n)
d
```

```r
plotAlloy = d %>% ggplot()+
  xlab(expression(paste("Thousandas of Cycles", " ", ("" %*% 10^3))))+
  ylab(expression(italic(p)))
plotAlloy + geom_point(aes(t, p))
```

```r
plotAlloy = plotAlloy + ylab("Standard Quantile") # y軸名を変更 
d = d %>% mutate(invNor = qnorm(p)) # 正規分布の逆関数
plotAlloy + geom_point(data=d, aes(x=t, y=invNor))
```

```r
d = d %>% mutate(log_t = log(t))
plotAlloy + geom_point(data=d, aes(x=log_t, y=invNor))
```

```r
d = d %>% mutate(invSev = log(-log(1-p)))
plotAlloy + geom_point(data=d, aes(x=t, y=invSev))
```

```r
plotAlloy + geom_point(data=d, aes(x=log_t, y=invSev))
```

```r
bind_rows(
  data.frame(model="Normal", x=d$t, y=d$invNor),
  data.frame(model="LogNormal", x=d$log_t, y=d$invNor),
  data.frame(model="SEV", x=d$t, y=d$invSev),
  data.frame(model="Weibull", x=d$log_t, y=d$invSev)
) %>%
  mutate(model = factor(model,
levels = c("Normal", "LogNormal", "SEV", "Weibull"))) %>%
ggplot() +
xlab(expression(paste("Thousandas of Cycles", " ", ("" %*% 10^3))))+
ylab(expression(italic(p)))+
geom_point(aes(x, y, col=model), show.legend = FALSE)+
facet_wrap(~model, scales="free")
```
